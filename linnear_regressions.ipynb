```python
from sklearn.model_selection import train_test_split
```

```python
x = df[['Time on App','Time on Website','Length of Membership']]
y = df[['Yearly Amount Spent']]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=101)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
```

```python
ln = LinearRegression()
```

```python
ln.fit(x_train, y_train)
```

```python
ln.coef_
```

```python
cdf =  pd.DataFrame(ln.coef_.T, x.columns, columns=['Coeff'])
print(cdf)
```

```python
predictions = ln.predict(x_test)
```

```python
predictions
```

```python
sns.scatterplot(x=predictions.flatten(), y=y_test['Yearly Amount Spent'])
plt.xlabel('Predictions')
plt.ylabel('True Yearly Amount Spent')
plt.title('Evaluation of our Linear Regression Model')
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import math
```

```python
print('Mean Absolute Eror :', mean_absolute_error(y_test, predictions))
print('Mean Squared Error :', mean_squared_error(y_test, predictions))
print('Root Mean Squared Error :', math.sqrt(mean_squared_error(y_test, predictions)))
```

```python
residuals = y_test - predictions
```

```python
residuals
```

```python
sns.displot(residuals, bins=50, kde=True)
```

```python
import pylab
import scipy.stats as stats

stats.probplot(residuals.iloc[:, 0], dist="norm", plot=pylab, fit=False)
pylab.show()
```
