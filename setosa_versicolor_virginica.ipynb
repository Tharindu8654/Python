```python
import pandas as pd

pd.plotting.scatter_matrix(pd.DataFrame(X_train, columns = iri_datasets.feature_names), c = y_train, figsize = (15,15), marker = 'o', hist_kwds = {'bins':20}, s = 60, alpha = .8)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 1)
```

```python
knn.fit(X_train, y_train)
```

```python
import numpy as np
X_new =np.array([[5,2.9,1,0.2]])
print("X new Shape: \n",X_new.shape)
```

```python
prediction = knn.predict(X_new)
print("Prediction :\n", prediction)
print("pPredicted target name:\n ", iri_datasets['target_names'][prediction])
```

```python
y_pred = knn.predict(X_test)
print("Test set predictions :\n", y_pred)
```

```python
from sklearn.metrics import accuracy_score
#print("Test set Score :{:.2f}".format(np.mean(y_pred ==y_test)))
print(accuracy_score(y_test, y_pred))
```
